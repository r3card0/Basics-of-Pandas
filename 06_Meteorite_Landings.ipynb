#### Como crear tipo de dato *category*

Se observa el contenido de las columnas y se puede determinar que el texto puede indicar categorias. En estadística, esto se le conoce como datos categóricos.

Se cuenta con un comando para indicar en el dataframe que el tipo de dato de estas columnas, son categorias. El comando *astype*

```python
# Crear una variable que guarde el nuevo tipo de dato en las columnas fall y nametype
df_meteorite[['fall','nametype']] = df_meteorite[['fall','nametype']].astype('category')
```

```python
# ver el cambio
df_meteorite.dtypes
```

Ahora vemos que en las columnas *fall* y *nametype*, su tipo de dato cambio de *object* a *category*, como se indicó en el código con el comando *astype*

Ahora, se puede emplear el comando unique, para conocer el o las categorias existentes en la columna *fall*

```python
df_meteorite['fall'].unique()
```

El resultado será que la columna *fall* tiene dos categorias: *Fell* y *Found*

```python
# conocer las frecuencias de las categorias Fell y Found en la columnas fall
df_meteorite['fall'].value_counts()
```

#### Tipos de Datos Dummies

Los tipos de dato dummies, sirven para convertir el formato de texto en formato numérico y esto en ocasiones facilita el análisis de un dataframe.

Aun no lo tengo muy claro, pero ya se que puedo convertir los contenidos de tipo texto de una columna a contenido de tipo numérico.

Para crear Datos Dummies, se usa el comando *get_dummies*

```python
# convertir formato de texto de una columna a formato numérico con get_dummies
pd.get_dummies(df_meteorite['fall'])
```

#### Crear nuevas columnas en el dataframe.

1. Crear nueva columna *Ones*. En este ejemplo se desea crear una nueva columna que contenga puros 1's dentro del dataframe

2. Crear las columnas Fell y Found con datos numéricos

```python
# 1. crear columna ones en el dataframe
df_meteorite['Ones'] = 1
df_meteorite
```

```python
# 2. Crear columnas Fell y Found
df_meteorite[['Fell','Found']] = pd.get_dummies(df_meteorite['fall'])
```

```python
# revisar el dataframe
df_meteorite
```

#### Crear una copia del Dataframe

Esto es muy útil para trabajar sobre la copia y no sobre el archivo original.

```python
# crear una copia del dataframe
# definir un nuevo dataframe usando variables
# especificar que la copia es profunda con el argumento deep=True
df_met_copy = df_meteorite.copy(deep=True)
```

```python
df_met_copy.head()
```

**Nota Importante**

Cuando se copia un dataframe en una variable, lo que se sucede es que la variable usa la ubicación del archivo origina como referencia. Esto quiere decir que si se modifica la copia, se modifica el original.

Para evitar eso, el argumento **deep=True**, genera una copia exacta del dataframe, permitiendo trabajar sobre la copia sin modificar el original.

Se recomienda mantener el archivo original intacto y trabajar sobre la copia.
