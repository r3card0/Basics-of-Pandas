## II Crear Dataframe

1. Crear Diccionario
2. Convertir el diccionario a dataframe
3. Guardar el directorio donde se alamcenaran los archivos

```python
#1. Crear un diccionario
dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
```

```python
#2.convertir el diccionario a dataframe
df = pd.DataFrame(dict_data)
df
```

```python
#3. Guardar el directorio donde se alamcenaran los archivos
#Guardar dataframe en Google Drive
#dir_pandas = "/content/drive/'My Drive'/'Colab Notebooks'/Platzi_courses_data/db/{}") #.format('test.csv')
dir_pandas = "/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/{}" #).format('test.csv')
dir_pandas
```

## III Crear archivo JSON a partir del dataframe

```python
# crear el archivo JSON
df.to_json(dir_pandas.format('test.json'))
```
