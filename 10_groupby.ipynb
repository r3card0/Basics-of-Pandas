```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
import seaborn as sns
df = sns.load_dataset('diamonds')
df.head()
```

```python
df.groupby('cut').mean()
```

```python
df.groupby('cut').median()
```

```python
df.groupby('cut')['carat'].count()
```

```python
df.groupby('cut')['carat'].max()
```

```python
df.groupby('cut')['carat'].min()
```

```python
for i in [0,1,2,3,'hola']:
  print(i)
```

```python
# agrupar por un parámetro
for key_group, group in df.groupby('cut'):
  grouped_price = group['price'].mean()

  print('Cut: {}, Price: {} \n'.format(key_group,grouped_price))
```

```python
# agrupar por varios parámetros
df.groupby(['cut','color'])['price'].mean()
```

```python
# convertir en dataframe con el comando to_frame()
df.groupby(['cut','color'])['price'].mean().to_frame()
# es un dataframe con mútiples índices
```

```python
# agrupar por corte y color y mostrar el precio mínimo, máximo y promedio
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,max])
```

```python
def mean_kilo(x):
  return np.mean(x)/1000
```

```python
# agrupar por corte y color y mostrar el precio mínimo, máximo, promedio y la función mean_kilo
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,max,mean_kilo]).head(10)
```

```python
dict_agg = {'carat':[min,max],'price':[np.mean,mean_kilo]}
dict_agg
```

```python
df.groupby(['cut','color']).aggregate(dict_agg)
```

## Crear filtros

```python
def f_filter(x):
  return mean_kilo(x['price'])>4
```

```python
df.groupby('cut').filter(f_filter).head()
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```
