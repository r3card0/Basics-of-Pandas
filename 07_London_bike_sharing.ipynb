```python
# crear una copia del dataframe
# definir un nuevo dataframe usando variables
# especificar que la copia es profunda con el argumento deep=True
df_londonb = df_london_bike.copy(deep=True)
```

## Cambiar el formato de fechas

```python
# almacenar en el dataframe el tipo de fato fecha
df_londonb['timestamp'] = pd.to_datetime(df_londonb['timestamp'])
df_londonb['timestamp']
```

**Nota 1**

Cuando el archivo original contiene una columna de fecha y esta tiene un formato relativamente organizado como el visto en *timestamp*, pandas, es capaz de reconocerlo y reliza el cambio son la necesidad de la línea de codigo: *format = '%m/%d/%Y %H:%M:%S %p'*

```python
# ver los cambios en dtypes
df_londonb.dtypes
```

```python
df_londonb
```

## Uso de funciones

### Funciones def simples

```python
# creacion de la funcion para convertir de grados Celcius a grados Farenheit
# aplicar la formula en la funcion: F = (9C/5)+32
def farenheit(celcius):
  y = ((9*celcius)/5) + 32
  return y
```

```python
# aplicacion de la funcion
farenheit(15)
```

### Funciones def complejas

```python
# crear un rango de valores de temperaturas, con el comando arange()
np.arange(-5,25)
```

```python
# conocer la cantidad de valores del range
np.arange(-5,25).shape
```

```python
# Calcular los grados Farenheit a partir del rango de valores
farenheit(np.arange(-5,25))
```

#### Como aplicar las funciones al dataframe

```python
# calcular los grados farenheti en el dataframe
# aplicar la funcion farenheit en la columna de temperatura t1
df['t1'].apply(farenheit)
```

#### Crear funciones con argumentos

```python
# funciones con argumentos definidos
# calcular la proporción de hombres y mujeres de un grupo dado
def gender(total, men, women):
  men_total = total * men
  women_total = total * women
  return men_total,women_total
```

```python
#aplicar funcion y cambiar argumentos
gender(215,men=.15, women=.85)
```

```python
# aplicar funcion en dataframe definiendo argumentos
df['cnt'].apply(gender, args = (.45,.55))
```

```python
# otra forma de definir argumentos
df['cnt'].apply(gender, men=(.23), women=(.77))
```

### Funciones lambda

Son funciones para crear código de forma más rápida y sencilla, en una sola línea

```python
# convertir de grados Celcius a grados Farenheit
# aplicar la fórmula en la funcion: F = (9C/5)+32
df['t1'].apply(lambda celcius: (9 * celcius / 5) + 32)
```

```python
# aplicando una función a todo el dataframe
# calcular los promedios de las columnas
df.apply(lambda x: x.mean())
```

```python
# calcular los promedios de las filas del dataframe, con el comando axis
df.apply(lambda x: x.mean(),axis=1)
```

```python
# realizar operaciones entre columnas
# Calcular la diferencia entre las columnas de temperatura t1 y t2
df.apply(lambda x: x['t1'] - x['t2'], axis=1)
```

```python
# aplicar la función a todo el dataframe
# dividir entre 1000 todos los valores del dataframe
df.applymap(lambda x: x / 1000)
```
