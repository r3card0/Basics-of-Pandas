## Crear DataFrame a partir de la libreria Seaborn

Se crea una base de datos de un restaurante

```python
df = sns.load_dataset('tips')
df
```

## Agrupar por sex y time y calcular el promedio de total_bill

```python
df.groupby(['sex','time'])[['total_bill']].mean()
```

Borrar indices y hacerlos columnas

```python
df_gp = df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

## Aplicar tabla dinÃ¡mica

Hizo como un transpose :P

Pivot table calcula el promedio de Lunch para Male y Female y Dinner igual

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```python
df_gp.pivot_table(values='total_bill', index='sex',columns='time')
```

```python
# otra forma que calcula lo mismo promedio
df.pivot_table(values='total_bill', index='sex',columns='time')
```

### Ahora a calcular la mediana

```python
df.pivot_table(values='total_bill', index='sex',columns='time', aggfunc=np.median)
```

### Calcular mediana y desviaciÃ³n estandar std

```python
df.pivot_table(values='total_bill', index='sex',columns='time', aggfunc=[np.median,np.std])
```

### Deshace las categorias para convertiras en filas

```python
# 1. Guardar el codigo anterior en una variable
df_pivot = df.pivot_table(values='total_bill', index='sex',columns='time', aggfunc=[np.median,np.std])
df_pivot
```

```python
# 2. Deshacer las categorias y convertirlas en filas
df_pivot.unstack()
```

```python
# 3. Convertir a DataFrame
df_pivot.unstack().reset_index()
```
