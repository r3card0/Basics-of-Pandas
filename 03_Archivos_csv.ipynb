## II Generar archivo en formato .csv

1. Conocer como se pueden manejar las variables con datos tipo texto y usarlas a favor


Las variables de tipo texto se pueden reemplazar con los las llaves { }, con el fin de tener un espacio al cual asignarle valores. 

Supongamos que se tiene el valor 'Hola Mundo' y al usar las llaves { } y sustiturlas por 'Hola { }', posteriormente se puede asignar un valor deseado:

```python
# sustitucion de { } en una variable tipo texto
'Hola Mundo'
```

```python
# todo lo que se coloca dentro de format(), será incluido en el texto
'Hola {}'.format('Amigos')
```

## III Guardar los archivo generados en la carpeta de Google Drive

1. Importar las librerias de Pandas y numpy
2. Checar la versión de pandas en colab
3. Crear un diccionario de datos
4. converit el diccionario en dataframe
5. Guardar dataframe en Google Drive

```python
# 1. Importar las librerias de Pandas y numpy
import pandas as pd
import numpy as np
```

```python
#2. checar la versión de pandas
pd.__version__
```

```python
#3. Crear un diccionario
dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
```

```python
#4.convertir el diccionario a dataframe
df = pd.DataFrame(dict_data)
df
```

```python
#Guardar dataframe en Google Drive
# usar linea de codigo
#dir_pandas = "/content/drive/'My Drive'/'Colab Notebooks'/Platzi_courses_data/db/{}".format('test.csv')
dir_pandas = "/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/{}".format('test.csv')
dir_pandas
```

```python
#
df.to_csv(dir_pandas)
```

```python
# verificar la creación del archivo
%ls /content/drive/'My Drive'/'Colab Notebooks'/Platzi_courses_data/db/
```

```python
# crear el archivo sin índice
df.to_csv(dir_pandas, index = False)
```

## IV Leer el archivo .csv

1. Seleccionar la ruta del archivo y usar la funcion read_csv

```python
# usar la función read_csv
pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/test.csv')
# asignarle una variable al codigo anterior
df_read = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/test.csv')
df_read
```

El formato .csv es un formato que separa los datos con ",".

Hay otro tipo de formatos donde separan los datos con |

Se pueden cambiar los separadores con la función sep

```python
# ejemplo de uso de la función sep
df.to_csv(dir_pandas, sep = '|', index = False)
```

```python
# asi se ven los datos
df_read = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/test.csv')
df_read
```

```python
# para visualizar los datos, separados
df_read = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/test.csv',sep = '|')
df_read
```

La función sep, nos puede ayudar cuando recibamos archivos con formatos distintos al csv y que estan separados por otros simbolos.
