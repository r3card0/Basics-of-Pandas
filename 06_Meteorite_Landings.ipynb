## II Explorar el archivo Meterorite_Landings.csv

### II.1 Comandos para explorar el archivo

1. read_csv( ) 
2. head( )
3. tail( )
4. sample

```python
# 1. usar la función read_csv
pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/Meteorite_Landings.csv')
# asignarle una variable al codigo anterior
df_meteorite = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Platzi_courses_data/db/Meteorite_Landings.csv')
df_meteorite
```

```python
# 2. si es muy grande el archivo, se puede usar el comando .head(), para que muestre los primeros 5 registros
df_meteorite.head()
```

```python
# 2.1 o si se quieren ver los primeros 10, solo modificar el comando con argumentos
df_meteorite.head(10)
```

```python
# 3. si se requieren ver los ultimos registros, se usa el comando .tail()
df_meteorite.tail()
```

```python
# 3.1 de igual forma como head, en tail se pueden agregar argumentos
df_meteorite.tail(12)
```

```python
# 4. para ver registros aleatorios, se usa el comando .sample()
df_meteorite.sample()
```

```python
# 4.1 agregando argumentos a sample()
df_meteorite.sample(8)
```

## II.2 Comandos para examinar el tamaño de la base de datos.

1. shape
2. size
3. describe( )

```python
# shape, muestra el numero de registros(filas) y el número de columnas
df_meteorite.shape
```
