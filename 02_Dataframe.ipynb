```python
import numpy as np
```

```python
dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
df2 = pd.DataFrame(dict_data)
df2
```

```python
# Se puede filtrar la informacion con operadores
# por ejemplo filtrar el dataframe con las edade iguales o menores a 12
df2['edad'] <= 12
```

```python
df2[df2['edad'] <= 12]
```

```python
# se pueden combinar varias condiciones al mimo tiempo, separandolas por parentesis
# por ejemplo se requiere filtrar por personas menor a 12 años y que sean de México
df2[(df2['edad'] <= 12) & (df2['pais'] == 'mx')]
```

```python
# otra forma es usando querys
df2.query('edad <= 12')
```

```python
# comparación de valores entre dos columnas 
# por ejemplo filtrar la info de los estudiantes que mejoraron sus calificaciones en el Q2
# extraer algunos datos del renglón
df2[df2['Q2'] >= df2['Q1']]
```
