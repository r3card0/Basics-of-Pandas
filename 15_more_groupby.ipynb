## Crear DataFrame a partir de la libreria Seaborn

Se crea una base de datos de un restaurante

```python
df = sns.load_dataset('tips')
df
```

## Crear la columna ones para . . .

```python
df['ones'] = 1
df
```

## Agrupar por genero y por fumador

Se crea un dataset que indica:
1.  el numero de personas male que SI fuman
2. el numero de personas male que NO fuman
3.  el numero de personas female que SI fuman
4. el numero de personas female que NO fuman

```python
# Agrupar por las categorias sex y smoke y sumar el numero de registros con la columnas 'ones'
df_g = df.groupby(['sex','smoker'])[['ones']].sum()
df_g
```

## Calcular los porcentajes de fumadores

1. convertir los valores numéricos de la columna "ones" a porcentajes
2. Agrupar por el nivel del género. level=0
3. La variable "x" toma los valores numéricos

```python
df_g.groupby(level=0).apply(lambda x: x/x.sum() * 100)
```

## Categorizar valores contínuos a categóricos

```python
# crear 3 categorias usando bins
pd.cut(df['total_bill'],bins = 3).value_counts()
```

```python
# crear categorias por valores indicados por el usuario
pd.cut(df['total_bill'],bins = [3,18,35,60]).value_counts()
```

## Crear la categoría en el dataframe

```python
df['categoria'] = pd.cut(df['total_bill'], bins = [3,18,35,60])
df
```

## Hacer agrupaciones por time y categoria recien creada

```python
df.groupby(['time','categoria'])[['ones']].count()
```

### convertir a porcentajes

```python
df.groupby(['time','categoria'])[['ones']].count().groupby(level=0).apply(lambda x: x/x.sum() * 100)
```
