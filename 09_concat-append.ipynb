## Concatenar Matrices
Numpy puede concatenar matrices usando la función *concatenate*

### Concatenar Matrices con Numpy

```python
# unir las matrices usando la función concatenate
np.concatenate([x1,x2])
```

```python
# ver la dimensiones del objeto creado
np.concatenate([x1,x2]).shape
```

Se creo un objeto de 4 columnas con 5 valores cada una. En el anterior ejemplo se creo un matriz a lo largo del eje de las filas

#### Crear Matrices a lo largo de los ejes de las columnas

```python
np.concatenate([x1,x2], axis = 1)
```

```python
# ver la dimensiones del objeto creado
np.concatenate([x1,x2], axis = 1).shape
```

Se creo un objeto de 2 columnas con 10 valores cada una

### Concatenar Matrices con Pandas

```python
# Crear set de datos
# se toma la primera columna de x1
# de definen los índices
s1 = pd.Series(x1[0],index = ['a','b','c','d','e'])
s1
```

```python
# Crear una segunda serie de elementos
s2 = pd.Series(x2[0], index = ['c','d','e','f','g'])
s2
```

### Concatenar

```python
# concatenar a lo largo de las filas
pd.concat([s1,s2])
```

```python
# concatenar a lo largo de las columnas
pd.concat([s1,s2], axis=1)
```

**Nota** -> El anterior ejemplo coloca los calores haciendo coincidir los indices descritos anteriormente en ambos elementos y creanod NaN donde no existen coincidencias

### Concatenar matrices SIN respetar los índices

## Concatenar Dataframes

```python
# Crear dataframe 1
df1 = pd.DataFrame(np.random.rand(3,2) * 10, columns=['a','b'])
df1
```

```python
# Crear dataframe 2
df2 = pd.DataFrame(np.random.rand(3,2) * -1, columns=['a','b'], index = [2,3,4])
df2
```

```python
# concatenar dataframes a lo largo de las filas
pd.concat([df1,df2])
```

```python
# concatenar dataframes a lo largo de las columnas
pd.concat([df1,df2], axis = 1)
```

```python
# hacer énfasis del índice que se repite en ambos dataframes
pd.concat([df1,df2], axis = 1, join = 'inner')
# y se obtiene únicamente fila del indice que comparten ambos dataframes
```

### Concatenar dataframes SIN respetar los índices

## Juntar dataframes usando el comando *append*

```python
df1.append(df2)
```

```python
# Se puede pegar varias veces
df1.append(df2).append(df2)
```

```python
# traspuesta. Invertir el orden. Poner columnas como indices y poner indices como columnas
df1.T
```

```python
# hacer append en la traspuesta
df1.T.append(df2.T).T
```
