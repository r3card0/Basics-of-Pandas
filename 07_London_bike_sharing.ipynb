## Crear copia del dataset

```python
# crear una copia del dataframe
# definir un nuevo dataframe usando variables
# especificar que la copia es profunda con el argumento deep=True
df_londonb = df_london_bike.copy(deep=True)
```

```python
# dataset con el que se trabajará
df_londonb
```

## Cambiar el formato de fechas

```python
# almacenar en el dataframe el tipo de fato fecha
df_londonb['timestamp'] = pd.to_datetime(df_londonb['timestamp'])
df_londonb['timestamp']
```

**Nota 1**

Cuando el archivo original contiene una columna de fecha y esta tiene un formato relativamente organizado como el visto en *timestamp*, pandas, es capaz de reconocerlo y reliza el cambio son la necesidad de la línea de codigo: *format = '%m/%d/%Y %H:%M:%S %p'*

```python
# ver los cambios en dtypes
df_londonb.dtypes
```

```python
df_londonb
```

## Crear columna *hour*

```python
# se aplica cuando la columna ya tiene definido el formato tipo tiempo/fecha
df_londonb['hour'] = df_londonb['timestamp'].dt.hour
df_londonb['hour']
```

### Operaciones entre columnas

```python
# calcular la temperatura real y la sensación térmica
# tr = t1-t2
tr = df['t1'] - df['t2']
tr
```

```python
# operaciones entre columnas usando filtros, con el comando iloc()
# calcular la temperatura real únicamente de los indices par
# 1. extraer las filas de los indices par
df['t1'].iloc[::2]
```

```python
# 2. realizar resta
df['t1'].iloc[::2] - df['t2']
```

Aqui se observa que las filas impares, retornan el valor NaN

En este caso particular no aplica, pero cuando se requiera sustituir los valores NaN, se hace lo siguiente:

df['t1'].iloc[::2].sub(df['t2'], fill_value = 1000)

El 1000, es el valor que se le asignará a las celdas con el valor NaN

```python
# 2. realizar resta
tr_2 = df['t1'].iloc[::2] - df['t2'].iloc[::2]
tr_2
```

```python
# división
df['t1'] / df['t2']
```

```python
#algebra lineal, producto punto de una misma columna con el comando dot
df['t1'].dot(df['t1'])
# se multiplican entre si cada valor de cada celda y se suman todas ls celdas.
```
