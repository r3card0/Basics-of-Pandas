## Aplicar funciones

Convertir de EURO a USD

```python
# crear funcion de conversion
def mean_eur2usd(x):
  return np.mean(x)*1.12
```

```python
# prueba
mean_eur2usd([100,102])
```

```python
# Calcular el promedio de la cuenta y propina por genero en dolares
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
```

## Agregando más variables

```python
# Calcular el promedio de la cuenta y propina por genero y tiempo en dolares
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

### Cuando se usa el comando *apply*, se puede usar funciones definidas por nostros mismos y también por funciones de las librerias

```python
# Calcular la desviación estandar de la cuenta y propina por genero y tiempo en dolares
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std)
```

## Crear funciones *lambda*

```python
# Calcular el promedio de la cuenta y propina por genero y tiempo en dolares usando la funcion lambda
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

## Aplicar más de una función, usando el comando *aggregate*

```python
# Calcular el promedio, el valor máximo y mínimo de la cuenta y la propina por genero y tiempo en dolares usando aggregate para integrar una lista de funciones
#aggregate([lista de funciones])
## forma corta de escribir el comando aggregate -> agg
df.groupby(['sex','time'])[['total_bill','prct_tip']].aggregate([np.mean,np.max,np.min])
```

```python
## forma corta de escribir el comando aggregate -> agg
df.groupby(['sex','time'])[['total_bill','prct_tip']].agg([np.mean,np.max,np.min])
```

## Crear un diccionario de operaciones

1. Va a contener dos llaves
2. La primera llave es de las propinas. Y se calculara el valor máximo y el valor mínimo en los grupos de género y tiempo
3. La se segunda llave sera de la cuenta y se calculará el promedio en euros y el promedio en dólares

```python
# creación del diccionario
dict_agg = {'tip':[min,max],'total_bill':[np.mean,mean_eur2usd]}
dict_agg
```

```python
# Colocar el diccionario en aggregate
df.groupby(['sex','time'])[['total_bill','tip']].agg(dict_agg)
```

## Definir filtros

```python
# crear la funcion que extraiga los promedios de las cuentas mayores a 20
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

```python
df.groupby(['sex','time']).filter(f_filter)
```
