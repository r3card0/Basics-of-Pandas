## Crear copia del dataset

```python
# crear una copia del dataframe
# definir un nuevo dataframe usando variables
# especificar que la copia es profunda con el argumento deep=True
df_pop = df_population.copy(deep=True)
```

```python
# dataset con el que se trabajará
df_pop.head()
```

## Separar por comas los valores numéricos

```python
# separar por comas los valores numericos
pd.options.display.float_format = '{:,.1f}'.format
```

```python
df_pop.head()
```

```python
# extraer nombre de las columnas
df_pop.columns
```

## Como filtrar por *columnas*

```python
# filtrar la información por paises
# crear la variable que guardara el filtro
# Elegir la info de los siguientes paises: Aruba y Colombia
filtro = df_pop['Country Name'].isin(['Aruba','Colombia'])
filtro
```

```python
df_pop_fil = df_pop[filtro]
df_pop_fil
```

## Como crear *índices*

```python
# Convertir columnas en índices
df_pop_ind = df_pop_fil.set_index(['Country Name','Country Code'])
df_pop_ind
```
